# Transformaciones geométricas

Una transformación geométrica redistribuye  los pixeles de una imagen. Si $I[r,c]$ es la intensidad de la imagen $I$ en el pixel ubicado en la fila $r$ y en la columna $c$, entonces por medio de una transformación geométrica, $T$, el pixel pasa a las coordenadas $[r', c']$, en la imagen transformada $I_T$, de manera que se cumplen las siguientes igualdades: 

\begin{align}
    [r',c'] &= T(r,c)\\
    &\\
    I[r, c] &= I_T[T(r, c)]
\end{align}

Siendo $T$ una función de dos variables. 

## Ejemplos de transformaciones geométricas

### Traslación

Para una traslación, la transformación $T$ se define de la siguiente manera: 

$$T(r,c) = [r', c'] = [r+t_r, c+t_c]$$

siendo $[t_r, t_c]$ el vector de traslación constante. 

Observa que la transformación puede reescribirse matricialmente como 

$$
\left[ 
   \begin{array}{c}
       r' \\
       c' 
   \end{array}
\right] =
\left[ \begin{array}{cc}
1 & 0 \\
0 & 1
\end{array} \right]
%
\left[ \begin{array}{c}
r\\
c
\end{array} \right]
+
\left[ 
   \begin{array}{c}
       t_r \\
       t_c 
   \end{array}
\right]
$$

```python
# Traslación de una imagen
#el parámetro offset guarda las componentes de la traslación: [-t_r, -t_c]
g_tr = ndimage.interpolation.affine_transform(g,[[1,0],[0,1]],order=2,offset=[100,-150],cval=0.0)

plt.imshow(g_tr, cmap="gray")
plt.axis('off')
plt.show()
```

### Rotación

Para una rotación, con un ángulo $\theta$, la transformación $T$ se define de la siguiente manera: 

$$T(r,c) = [r', c'] = [r\cos{\theta}-c\sin{\theta}, r\sin{\theta}+c\cos{\theta}]$$

Observa que la transformación puede escribirse matricialmente como 

$$
\left[ 
   \begin{array}{c}
       r' \\
       c' 
   \end{array}
\right] =
\left[ \begin{array}{cc}
\cos{\theta} & -\sin{\theta} \\
\sin{\theta} & \cos{\theta}
\end{array} \right]
%
\left[ \begin{array}{c}
r\\
c
\end{array} \right]
+
\left[ 
   \begin{array}{c}
       0 \\
       0 
   \end{array}
\right]
$$

```python
a=15.0*np.pi/180.0 #el ángulo de rotación debe ingresarse en radianes
rot=np.array([[np.cos(a),np.sin(a)],[-np.sin(a),np.cos(a)]])
g_tr=ndimage.interpolation.affine_transform(g,rot,order=2,offset=[0, 0],cval=0.0)

plt.imshow(g_tr, cmap="gray")
plt.axis('off')
plt.show()
```

### Escalamiento

Para un escalamiento, con fator de escala $s$, la transformación $T$ se define de la siguiente manera: 

$$T(r,c) = [r', c'] = [s\cdot r, s\cdot c]$$

Observa que la transformación puede escribirse matricialmente como 

$$
\left[ 
   \begin{array}{c}
       r' \\
       c' 
   \end{array}
\right] =
\left[ \begin{array}{cc}
s & 0 \\
0 & s
\end{array} \right]
%
\left[ \begin{array}{c}
r\\
c
\end{array} \right]
+
\left[ 
   \begin{array}{c}
       0 \\
       0 
   \end{array}
\right]
$$

```python
#el factor de escala es de 2, s=s, su inverso es s=0.5
g_tr=ndimage.interpolation.affine_transform(g,[[0.5,0],[0,0.5]],order=2,offset=[0, 0],cval=0.0)
plt.imshow(g_tr, cmap="gray")
plt.axis('off')
plt.show()
```

# Alineamiento de imágenes

Dadas dos imágenes, $I_1$ e $I_2$, buscamos determinar la transformación $T$ que permita alinearlas:

$$I_2[r, c] = I_1[r', c'] = I_1[T(r, c)]$$

para todos los pixeles [r, c].

La transformación puede obtenere minimizando la función de error:

\begin{equation}
   E(T) = \sum_{r, c} \left(I_2[r, c]-I_1[T(r,c)]\right)^{2}
\end{equation}

La suma se extiende sobre todos los pixeles de la imagen $I_2$.


Por ejemplo, para una **traslación**:

\begin{equation}
   E(t_r, t_c) = \sum_{r, c} \left(I_2[r, c]-I_1[r+t_r, c+t_c]\right)^{2}
\end{equation}


Mientras que para una **rotación**:

\begin{equation}
   E(\theta) = \sum_{r, c} \left(I_2[r, c]-I_1[r\cos{\theta}+c\sin{\theta}, -r\sin{\theta}+c\cos{\theta}]\right)^2
\end{equation}


```python
# Como ejemplo, consideramos la imagen "g" y su versión girada "g_tr"

I1 = g

a=15.0*np.pi/180.0 #el ángulo de rotación debe ingresarse en radianes
rot=np.array([[np.cos(a),np.sin(a)],[-np.sin(a),np.cos(a)]])
g_tr=ndimage.interpolation.affine_transform(I1,rot,order=2,offset=[0, 0],cval=0.0)

I2 = g_tr

plt.subplot(121)
plt.imshow(I1, cmap="gray")
plt.axis('off')

plt.subplot(122)
plt.imshow(I2, cmap="gray")
plt.axis('off')
plt.show()
```

consideremos diferentes ángulos y midamos el error global

### Observaciones

+ La evaluación de la función de error es muy costosa computacionalmente
+ Se puede realizar una **estimación** de la función de error considerando solamente un pequeño conjunto de muestras sobre la imagen $I_2$, por ejemplo 100 pixeles muestreados aleatoria y uniformemente sobre la imagen.

Considera que $P$ es un conjunto de pixeles muestreados sobre la imagen $I_2$. La estimación del error se puede realizar sumando solamente sobre los pixeles muestreados, considera adempas que será necesario evaluar las intensidades correspondientes a los pixeles transformados sobre la imagen $I_1$, esto es, evaluar $I_1[r', c']$

En la siguiente imagen ejemplificamos la evalaución para una traslación:


```python
# t es el vector de traslación
# P es el conjunto de coordenadas muestreadas
def EvalErrorSamps(I1, I2, P, t):
    err = 0
    (ren, col) = I1.shape
    for p in P:
        
        r = p[0]
        c = p[1]
        
        #coordenadas transformadas
        rt = r + t[0]
        ct = c + t[1]
        
        #busca las coordenadas del pixel más cercano
        rt = int(rt+0.5)
        ct = int(ct+0.5)
        
        if rt >= 0 and rt < ren and ct >= 0 and ct < col:
            err += (I2[r, c]-I1[rt, ct])**2
    return err
```

```python
#Construimos un par de imágenes para probar

# Como ejemplo, consideramos la imagen "g" y su versión trasladada "g_tr"

I1 = g

t=[100, 100]
rot=np.array([[1, 0],[0, 1]])

g_tr=ndimage.interpolation.affine_transform(I1,rot,order=2,offset=t,cval=0.0)

I2 = g_tr

plt.subplot(121)
plt.imshow(I1, cmap="gray")
plt.axis('off')

plt.subplot(122)
plt.imshow(I2, cmap="gray")
plt.axis('off')
plt.show()
```

```python
#Muestreamos la imagen

from random import randint

(ren, col) = I2.shape

n = 100 #número de muestras
P = []
for i in range(n):
    rs = randint(0, ren-1)
    cs = randint(0, col-1)
    
    P.append([rs, cs])

print P
```

Medimos el error para diferentes vectores de traslación
