# T2 Filtros derivadores

```python
#algunas librerías
from scipy import misc
from scipy import ndimage
from scipy import signal
import matplotlib.pylab as plt
import numpy as np
import matplotlib.pylab as plt
import numpy as np

import math
```

```python
def muestra_imagen(I, msg ='', show=True):
    plt.imshow(I, cmap="gray")
    m = np.min(I)
    mx = np.max(I)
    plt.xlabel(msg)
    if show:
        plt.show()
```

## Valores

```python
# Tamaño de la imagen
ren, col = 1000, 1000

# Valor de k
k = 1
```

## Se crea imagen
Dada una funcion creamos una imagen base

```python
img_c = np.zeros((ren, col))
for c in range(col):
    img_c[:, c] = 128 + 128*math.sin((2*math.pi*k*c) / 1000.0)

muestra_imagen(img_c)
```

## C

```python
img_c_derivadaUno, img_c_derivadaDos = np.zeros((ren, col)), np.zeros((ren, col))

# Derivadas
for c in range(col):
    img_c_derivadaUno[:, c] = ((32 * math.pi * k)/125.0) * math.cos((math.pi*k*c)/500.0)
    img_c_derivadaDos[:, c] = -((8 * (math.pi**2) * (k**2))/15625.0) * math.sin((math.pi*k*c)/500.0)
```

```python
# Filtros derivadores
c_kernelDerivadaUno, c_kernelDerivadaDos = np.array([[1, 0, -1]]), np.array([[1, -2, 1]])

Ix  = signal.convolve2d(img_c, c_kernelDerivadaUno, boundary='symm')
Ixx = signal.convolve2d(img_c, c_kernelDerivadaDos, boundary='symm')

# plt.plot(Ixx[600,10:-10])
```

```python
# Mostramos y comparamos los resultados
plt.figure(figsize=(17,17))
plt.subplot(131)
plt.imshow(img_c, cmap='gray')
plt.xlabel('Imagen original')
plt.subplot(132)
plt.imshow(img_c_derivadaUno, cmap='gray')
plt.xlabel('Primera derivada')
plt.subplot(133)
plt.imshow(img_c_derivadaDos, cmap='gray')
plt.xlabel('Segunda derivada')
plt.show()

plt.figure(figsize=(17,17))
plt.subplot(131)
plt.imshow(img_c, cmap='gray')
plt.xlabel('Imagen original')
plt.subplot(132)
plt.imshow(Ix, cmap='gray')
plt.xlabel('Primera derivada kernel')
plt.subplot(133)
plt.imshow(Ixx, cmap='gray')
plt.xlabel('Segunda derivada kernel')
plt.show()
```

## R

```python
img_r = np.zeros((ren, col))
for r in range(ren):
    img_r[r, :] = 128 + 128*math.sin((2*math.pi*k*r) / 1000.0)

muestra_imagen(img_r)
```

```python
img_r_derivadaUno, img_r_derivadaDos = np.zeros((ren, col)), np.zeros((ren, col))

# Derivadas
for r in range(ren):
    img_r_derivadaUno[r, :] = ((32 * math.pi * k)/125.0) * math.cos((math.pi*k*r)/500.0)
    img_r_derivadaDos[r, :] = -((8 * (math.pi**2) * (k**2))/15625.0) * math.sin((math.pi*k*r)/500.0)
```

```python
# Filtros derivadores
r_kernelDerivadaUno, r_kernelDerivadaDos = np.array([[1],[0],[-1]]), np.array([[1], [-2], [1]])

Iy  = signal.convolve2d(img_r, r_kernelDerivadaUno, boundary='symm')
Iyy = signal.convolve2d(img_r, r_kernelDerivadaDos, boundary='symm')
```

```python
# Mostramos y comparamos los resultados

plt.figure(figsize=(17,17))
plt.subplot(131)
plt.imshow(img_r, cmap='gray')
plt.xlabel('Imagen original')
plt.subplot(132)
plt.imshow(img_r_derivadaUno, cmap='gray')
plt.xlabel('Primera derivada')
plt.subplot(133)
plt.imshow(img_r_derivadaDos, cmap='gray')
plt.xlabel('Segunda derivada')
plt.show()

plt.figure(figsize=(17,17))
plt.subplot(131)
plt.imshow(img_r, cmap='gray')
plt.xlabel('Imagen original')
plt.subplot(132)
plt.imshow(Iy, cmap='gray')
plt.xlabel('Primera derivada kernel')
plt.subplot(133)
plt.imshow(Iyy, cmap='gray')
plt.xlabel('Segunda derivada kernel')
plt.show()
```
