```python
import numpy as np
import matplotlib.pyplot as plt
import cv2
```

```python
def muestra(im1, im2):
    plt.subplot(121)
    plt.imshow(im1, cmap="gray")
    plt.axis("off")
    plt.subplot(122)
    plt.imshow(im2, cmap="gray")
    plt.axis("off")
    plt.show()    
```

```python
im = cv2.imread("Sunflowers.jpg",0)
plt.imshow(im, cmap="gray")
plt.axis("off")
plt.show()
```

# Traslación

```python
t = [30, 40]
H = np.array([[1, 0, t[0]], [0, 1, t[1]], [0, 0, 1]], dtype='f')
H
```

```python
imt = cv2.warpPerspective(im,H,(im.shape[1],im.shape[0]))
muestra(im,imt)
```

# Escalamiento

```python
H = np.array([[0.5, 0, 0], [0, 2, 0], [0, 0, 1]], dtype='f')
H
```

```python
imt = cv2.warpPerspective(im,H,(im.shape[1]/2,2*im.shape[0]))
muestra(im,imt)
```

# Rotación

```python
th = -45*np.pi/180.0
H = np.array([[np.cos(th), -np.sin(th), 0], [np.sin(th), np.cos(th), 0], [0, 0, 1]], dtype='f')
H
```

```python
imt = cv2.warpPerspective(im,H,(im.shape[1],im.shape[0]))
muestra(im,imt)
```

# Cizallamiento

```python
H = np.array([[1, 0.3, 0], [0, 1, 0], [0, 0, 1]], dtype='f')
H
```

```python
imt = cv2.warpPerspective(im,H,(im.shape[1],im.shape[0]))
muestra(im,imt)
```

# Proyectiva

```python
H = np.array([[1, 0, 0], [0, 1, 0], [0.001, -0.001, 1]], dtype='f')
H
```

```python
imt = cv2.warpPerspective(im,H,(im.shape[1],im.shape[0]))
muestra(im,imt)
```
