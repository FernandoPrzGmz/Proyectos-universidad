```python
from sklearn.cluster import KMeans
from sklearn import datasets
import numpy as np
import matplotlib.pyplot as plt
from scipy import misc
from scipy import ndimage
```

```python
nClusters = 2

I = misc.imread("homer.jpg")
Icopy = np.copy(I)
```

```python
R,  G,  B  = Icopy[:,:,0],       Icopy[:,:,1],       Icopy[:,:,2]
_R, _G, _B = R.reshape((-1, 1)), G.reshape((-1, 1)), B.reshape((-1, 1))
 
X = np.concatenate( (_R, _G, _B), axis=1 )
X
```

```python
kMeans = KMeans(n_clusters = nClusters)
kMeans.fit(X)

centroids, labels = kMeans.cluster_centers_, kMeans.labels_
```

```python
for i in range(_R.shape[0]):
  _R[i], _G[i], _B[i] = centroids[labels  [i]][0], centroids[labels [i]][1], centroids[labels  [i]][2] 
_R.shape, _G.shape, _B.shape = R.shape, G.shape, B.shape 
_R, _G, _B = _R[:, :, np.newaxis], _G[:, :, np.newaxis], _B[:, :, np.newaxis]

Img = np.concatenate((_R, _G, _B), axis=2)
Img
```

```python
# Mostrar cambios
plt.figure(figsize=(12,12))
plt.subplot(121)
I_m = np.min(I)
I_mx = np.max(I)
plt.xlabel("["+str(I_m)+", "+str(I_mx)+"]")
plt.imshow(I)

plt.subplot(122)
Img_m = np.min(Img)
Img_mx = np.max(Img)
plt.xlabel("["+str(Img_m)+", "+str(Img_mx)+"]")
plt.imshow(Img)
plt.show()
```
